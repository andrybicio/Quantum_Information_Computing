```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = np.genfromtxt('O1.dat')
data2 = np.genfromtxt('O2.dat')
data3 = np.genfromtxt('O3.dat')
datafast = np.genfromtxt('Ofast.dat')
```

```python
size_matrix = data1[:,0]
definition_funct_times1 = data1[:,1]
transposed_funct_times1 = data1[:,2]
intrinsic_funct_timesfast1 = data1[:,3]
```

```python
definition_funct_times2 = data2[:,1]
transposed_funct_times2 = data2[:,2]
intrinsic_funct_timesfast2 = data2[:,3]
```

```python
definition_funct_times3 = data3[:,1]
transposed_funct_times3 = data3[:,2]
intrinsic_funct_timesfast3 = data3[:,3]
```

```python
definition_funct_timesfast = datafast[:,1]
transposed_funct_timesfast = datafast[:,2]
intrinsic_funct_timesfastfast = datafast[:,3]
```

```python
fig, ax = plt.subplots( 2,2, figsize = (12,8) , sharex = True, sharey = True)

ax[0,0].plot(size_matrix, definition_funct_times1, label = 'def')
ax[0,0].plot(size_matrix, transposed_funct_times1, label = 'transp')
ax[0,0].plot(size_matrix, intrinsic_funct_timesfast1, label = 'intrins')
ax[0,0].set_xlabel('size of square matrix')
ax[0,0].set_ylabel('time [s]')
ax[0,0].set_yscale('log')
ax[0,0].set_title("O1 optimization flag")
ax[0,0].legend(title = "O1")

ax[0,1].plot(size_matrix, definition_funct_times2, label = 'def')
ax[0,1].plot(size_matrix, transposed_funct_times2, label = 'transp')
ax[0,1].plot(size_matrix, intrinsic_funct_timesfast2, label = 'intrins')
ax[0,1].set_xlabel('size of square matrix')
ax[0,1].set_ylabel('time [s]')
ax[0,1].set_yscale('log')
ax[0,1].set_title("O2 optimization flag")
ax[0,1].legend(title = "O2")

ax[1,0].plot(size_matrix, definition_funct_times3, label = 'def')
ax[1,0].plot(size_matrix, transposed_funct_times3, label = 'transp')
ax[1,0].plot(size_matrix, intrinsic_funct_timesfast3, label = 'intrins')
ax[1,0].set_xlabel('size of square matrix')
ax[1,0].set_ylabel('time [s]')
ax[1,0].set_yscale('log')
ax[1,0].set_title("O3 optimization flag")
ax[1,0].legend(title = "O3")

ax[1,1].plot(size_matrix, definition_funct_timesfast, label = 'def')
ax[1,1].plot(size_matrix, transposed_funct_timesfast, label = 'transp')
ax[1,1].plot(size_matrix, intrinsic_funct_timesfastfast, label = 'intrins')
ax[1,1].set_xlabel('size of square matrix')
ax[1,1].set_ylabel('time [s]')
ax[1,1].set_yscale('log')
ax[1,1].set_title("O4 optimization flag")
ax[1,1].legend(title = "Ofast")

plt.savefig('Performances')
plt.show()
```
